# PIE CHART

```python
import matplotlib.pyplot as plt
classes = ["Python", "R", "ML","AI", "DS"]
class1_students=[45,15,35,25,30]
plt.pie([1])
plt.show()
```

```python

classes = ["Python", "R", "ML","AI", "DS"]
class1_students=[45,15,35,25,30]
plt.pie(class1_students,labels=classes,autopct ="%0.1f%%")
plt.legend()
plt.show()
```
