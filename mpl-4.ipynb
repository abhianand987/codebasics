```python
##plt.figure(figsize=(16,9))
x = df["Rating"]
y = df["Reviews"]
plt.scatter(x,y)
plt.title("Scatter Plot of ratings and reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
x = df["Rating"]

y = df["Reviews"]
plt.scatter(x,y,c='r',marker='2')
plt.title("Scatter Plot of ratings and reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
x = df["Rating"]

y = df["Reviews"]
plt.scatter(x,y,c='m',marker='o')
plt.scatter(x,df["Installs"],c='r',marker='o')
plt.title("Scatter Plot of ratings and reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```
