# HEATMAP
## Used to graphhically represent 2d data in the form of colour

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
df = pd.read_csv("E:\DataSets\Who_is_responsible_for_global_warming.csv")
df
```

```python
df2 = df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
df2.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2)
```

```python
##plt.figure(figsize=(16,9))
sns.heatmap(df2,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2,vmin=0,vmax=21,cmap="coolwarm")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2,vmin=0,vmax=21,cmap="coolwarm",annot=True)
```

# Corelation

```python
df2.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2.corr(),annot=True)
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(df2.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color="k",labelsize=10,labelcolor="k")
```
