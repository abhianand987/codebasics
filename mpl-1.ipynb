# matplotlib()=2d and 3d plotting python library

-It is used when you have data in bulk and aapka dimag chakra jyega dekh kar

-Histogram,-BarChart,Scatter plot,pie plot etc


```python
from matplotlib import pyplot as plt
            ##OR
import matplotlib.pyplot as plt    
```

# LINEPLOT = To Plot The LINE

```python
days =list(range(1,16))
print(days)
temperature =[30,34,38,41,41.5,42,39,29,34,39,38,40,41,43,44]
```

```python
plt.plot(days,temperature)##plt.plot(x,y)
plt.title('Delhi Temp')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.show()
```

```python
plt.plot(days,temperature)      ##plt.plot(x,y)
plt.axis(0,20,25,50)     ###(xmin,xmax,ymin,ymax) ##to give the starting point of co-ordinate system  
plt.title('Delhi Temp')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.show
```

```python
### above error bcoz we have to give paramter of axis in list input BC
plt.plot(days,temperature)      ##plt.plot(x,y)
plt.axis([0,30,0,50])     ###(xmin,xmax,ymin,ymax)
plt.title('Delhi Temp')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.show
```

```python
plt.plot(days,temperature,color='m',marker='o',linestyle='--',linewidth=3,markersize=8)      ##plt.plot(x,y)
plt.title('Delhi Temp')
plt.xlabel('Days')
plt.ylabel('Temp')
plt.show
```

```python
plt.plot(days,temperature,'go--',linewidth=3,markersize=8)      ###shortcut of not using color ,marker and linestyle.
plt.title('Delhi Temp',fontsize=15)
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temp',fontsize=12)
plt.legend(['Temp line'],loc=4)     ##used to label the graph bhai  ##or use label parameter in plot()
plt.show
```

```python
plt.plot(days,temperature,'go--',linewidth=3,markersize=8,label='Temp Line')      ###shortcut of not using color ,marker and linestyle.
plt.title('Delhi Temp',fontsize=15)
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temp',fontsize=12)
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
from matplotlib import style
days =list(range(1,16))
delhi_temp =[30,34,38,41,41.5,42,39,29,34,39,38,40,41,43,44]
mumbai_temp =[25,17,31,24,16,21,29,32,27,26,28,34,27,29,30]
style.use("ggplot")
plt.plot(days,delhi_temp,'bo--',linewidth=3,markersize=8,label='Delhi Temp Line')
plt.plot(days,mumbai_temp,'mo--',linewidth=3,markersize=8,label='Mumbai Temp Line') 

##if wanna add multiple plots then use plt.plot() multiple times mere bhai

plt.title('Delhi Mumbai Temp',fontsize=15)
plt.xlabel('Days',fontsize=12)
plt.ylabel('Temp',fontsize=12)
plt.legend(loc=4)
plt.grid(color='w',linestyle='-',linewidth=2)
plt.show()
```
