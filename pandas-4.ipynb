```python
df = pd.read_csv('E:\\Datasets\\Fortune_10.csv',na_filter=True)  ###na_filter true me NaN print and False me blank space
```

```python
df1 = pd.read_csv("E:\\Datasets\\Fortune_5.csv")
df1
```

```python
df1.dropna(axis =0)     ###Jis row NaN hoti hai usko drop kr deta hai ye how=any mtlb 1 bhi NaN to drop and all mtlb agar sarra
                        ###NaN tbhi drop thresh=any_int(eg-2) mtlb agar min 2 non-missing/not-null values hogi to print kro
                        ###subset use krke un specific columns ke un rows ko drop kr skte jisme NaN hai
                        ###inplace=True me usi dataframe ko update krta and naya df ni create krta hai
```

```python
df1.dropna(how ='all')  
```

```python
df1.fillna(123)###NaN ko is value se fill krdo Bc
```

```python
df1.fillna(0,limit=5)   ###if kisi specific col ka NaN ko change krna to we use dictionaray
                       ### method=bfill/ffill me backfill/forwardfill se value bhar dega NaN me BC
                        ###limit=5 mtlb har column me sirf 5 hi NaN ko fill krega BC
```
