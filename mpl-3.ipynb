### Pllotting BAR Chart

import matplotlib.pyplot as plt

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
style.use("ggplot")
plt.bar(classes,class1_students,width=0.6,color='m',label="class1 students",linestyle='--')
plt.title("Bar chart of IAIP Class")
plt.xlabel("classes")
plt.ylabel("Number of students",color='g',fontsize=15)
plt.legend()
```

```python
style.use("ggplot")
plt.bar(classes,class1_students,width=0.2,color='m',label="class1 students",linestyle='--')
plt.bar(classes,class2_students,width=0.2,color='k',label="class2 students",linestyle='--')
plt.bar(classes,class3_students,width=0.2,color='g',label="class3 students",linestyle='--')
plt.title("Bar chart of IAIP Class")
plt.xlabel("classes")
plt.ylabel("Number of students",color='g',fontsize=15)
plt.legend()
```

```python
style.use("ggplot")
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index,class1_students,width,color='m',label="class1 students",linestyle='--')
plt.bar(classes_index+width,class2_students,width,color='k',label="class2 students",linestyle='--')
plt.bar(classes_index+ 2*width,class3_students,width,color='g',label="class3 students",linestyle='--')

plt.xticks(classes_index+width,classes)

plt.title("Bar chart of IAIP Class")
plt.xlabel("classes")
plt.ylabel("Number of students",color='g',fontsize=15)
plt.legend()
```
