# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
df = sns.load_dataset("tips")
df
```

```python
sns.distplot(df["size"])
```

```python
sns.distplot(df["size"],hist=False)
```

```python
sns.distplot(df["total_bill"])
```

```python
sns.distplot(df["total_bill"],fit=norm)
```

```python
sns.distplot(df["total_bill"],fit=norm,kde=False,label="Total bill")
plt.legend()
plt.title("Histogram of total bill")
```
