# groupby() = split data into group based on some criteria

```python
import pandas as pd
df = pd.read_csv('E:\\Datasets\\weather_by_cities.csv',parse_dates =['day'])
df
```

```python
g = df.groupby('city')     ###isse hoga kya city key bn jyga and uske corresponding dataframe value bnegi bc
g
```

```python
for city,city_df in g:     ###city_df provided by groupby() bhai
    print(city)
    print(city_df)
```

```python
g.get_group('mumbai')
```

```python
g.max()
```

```python
g.mean()
```

```python
g.describe()
```

## concat() = used to join two or more df

```python
import pandas as pd
df1 = pd.DataFrame({'ID':[1,2,3,4],
    'name':['a','b','c','d'],
    'class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
    'name':['e','f','g','h'],
    'class':[9,10,11,12]})
df2
```

```python
type(df2)
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df3 = pd.concat([df1,df2],ignore_index=True)
df3
```

```python
df3 = pd.concat([df1,df2],axis=1)
df3
```

```python
df3 = pd.concat([df1,df2],keys=['first df','second df'])
df3
```

## join()

```python
dfr1 = pd.DataFrame({'A':[1,2,3],'B':[10,20,30]})
dfr1
```

```python
dfr2 = pd.DataFrame({'C':[4,5,6],'D':[40,50,60]})
dfr2
```

```python
display(dfr1,dfr2)
```

```python
pd.concat([dfr1,dfr2])
```

```python
dfr1.join(dfr2)
```

# merge()

```python
x1 =pd.DataFrame({'ID':[1,2,3],'Class':[8,9,10]})
x1
```

```python
x2 =pd.DataFrame({'ID':[1,2,4],'Name':['a','b','c']})
x2
```

```python
pd.merge(x1,x2,on='ID')
```

```python
pd.merge(x1,x2,on='ID',how='outer',indicator=True) 
```
