```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))  ###syntax(ll,ul,(Tuple input))
py_student_age = np.random.randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML STUDENT AGE HISTOGRAM")
plt.xlabel('Student age range')
plt.ylabel('No of students')
plt.show()
```

```python
plt.hist(py_student_age,color="m",bins=[15,20,25,30,35,40],rwidth=0.8,orientation='vertical')
plt.title("ML STUDENT AGE")
plt.xlabel("Student age range")
plt.ylabel("No of students")
plt.legend(["ML"])
plt.show()
```

```python
plt.hist(py_student_age,color="m",bins=[15,20,25,30,35,40],rwidth=0.8,orientation='vertical', label='py age')
plt.hist(ml_student_age,color="y",bins=[15,20,25,30,35,40,45],rwidth=0.8,orientation='vertical',label='ml age')
plt.title("ML STUDENT AGE")
plt.xlabel("Student age range")
plt.ylabel("No of students")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.hist([py_student_age,ml_student_age],color=["m",'y'],bins=[15,20,25,30,35,40],rwidth=0.8,orientation='vertical', label=['py age','ml_age'])
##plt.hist(,color="y",bins=[15,20,25,30,35,40,45],rwidth=0.8,orientation='vertical',label='ml age')
plt.title("ML STUDENT AGE")
plt.xlabel("Student age range")
plt.ylabel("No of students")
plt.legend()
plt.show()
```
