```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
df = sns.load_dataset("tips")
df
```

```python
sns.barplot(x=df.day,y=df.total_bill)
```

```python
sns.barplot(x=df.day,y=df.total_bill,hue=df.sex)
```

# Scatter Plot

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.scatterplot(x="age",y="fare",data=df)
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who")
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who",size="who")
```

```python
sns.scatterplot(x="who",y="fare",data=df,hue="alive",style="alive")
```
